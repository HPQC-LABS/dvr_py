First, we'll test the sinc function DVR on a simple harmonic oscillator potential.

```python
d = dvr_1d.SincDVR(43, 20)
d.sho_test(precision=12)
```

Let's try the same potential but with a Hermite basis. We see that since the Hermite polynomials are the exact solutions to the SHO problem, we can use `npts == num_eigs`. The eigenvectors will look horrible but the eigenvalues will be very accurate.

```python
d = dvr_1d.HermiteDVR(npts=5)
d.sho_test(precision=14)
```

```python
d = dvr_1d.SincDVR(npts=1000, L=20)
d.square_well_test(precision=6)
d.inf_square_well_test(precision=6)
```

```python
d = dvr_1d.SineDVR(npts=100, xmin=-20, xmax=20)
d.sho_test(precision=13)
d.square_well_test(precision=6)
d.inf_square_well_test(precision=6)
```

## Testing Area

```python
d1d = dvr_1d.HermiteDVR(npts = 10)
d2d = dvr_2d.DVR(dvr1d=d1d)
d2d.sho_test(num_eigs=10)
```

```python
vF = dvr_2d.VFactory()
V = vF.sho()
Vplt = V(d2d.xy)
xy = d2d.xy.reshape((d2d.dvr1d.npts, d2d.dvr1d.npts, 2))
Vplt = Vplt.reshape((d2d.dvr1d.npts, d2d.dvr1d.npts))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xyplt[:,:,0], xyplt[:,:,1], Vplt)
```
