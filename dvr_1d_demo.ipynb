## Testing Area

```python
npts = 4
d = dvr_1d.SincDVR(npts=npts, L=5)
d.t()
eye = np.eye(npts)

vF = dvr_1d.VFactory()
V = vF.square_well()
V_matrix = np.diag(V(d.x))
V_matrix

T = d.t()
```

```python
%%timeit
Ham = np.empty((npts**2, npts**2))
Vma = np.empty((npts**2, npts**2))
ii = -1
for ix in range(npts):
    for iy in range(npts):
        ii += 1
        jj = -1
        for jx in range(npts):
            for jy in range(npts):
                jj += 1
                Ham[ii, jj] = d.t()[ix, jx] * eye[iy, jy] + eye[ix, jx] * d.t()[iy, jy]
                Vma[ii, jj] = eye[ix, jx] * eye[iy, jy] * V_matrix[ix, iy]
```

```python
%%timeit
Ham_test = np.kron(eye, d.t()) + np.kron(d.t(), eye)
Vma_test = np.diag(np.diag(np.outer(eye,V_matrix)))
```

```python
plt.spy(Ham_test)
```

```python
print np.allclose(Ham, Ham_test, rtol=1e-16, atol=1e-16)
print np.allclose(Vma, Vma_test, rtol=1e-16, atol=1e-16)
```

```python
def cartesian_product(arrays):
    broadcastable = np.ix_(*arrays)
    broadcasted = np.broadcast_arrays(*broadcastable)
    rows, cols = reduce(np.multiply, broadcasted[0].shape), len(broadcasted)
    out = np.empty(rows * cols, dtype=broadcasted[0].dtype)
    start, end = 0, rows
    for a in broadcasted:
        out[start:end] = a.reshape(-1)
        start, end = end, end + rows
    return out.reshape(cols, rows).T
```
