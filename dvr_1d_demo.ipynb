```python
import dvr_1d as dvr
```

```python
import numpy as np
import scipy.sparse.linalg as sla
```

```python
V = dvr.square_well_factory(depth=10e30, width=20.)
dvr.test_potential(V, -100., 100., npts=2000, num_eigs=5, ymin=-0.25, ymax=1.5)
d = dvr.sincDVR1D(2000, 200.)
E, U = sla.eigsh(d.h(V), k = 5, which='LM', sigma=V(d.x).min())
print 'The first {n:d} energies are:\n'.format(n=5), E[:5]
dvr.plotter(d.x, V, E, U, nplot=5, 
            ymin=-0.25, 
            ymax=1.5, 
            doshow=True)
```
